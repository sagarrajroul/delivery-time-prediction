```python
import pandas as pd
import numpy as np

# Assuming you have a DataFrame named 'df' with the provided data.

# Function to calculate distance using Haversine formula
def haversine(lat1, lon1, lat2, lon2):
    R = 6371.0  # Radius of the Earth in kilometers
    lat1_rad = np.radians(lat1)
    lon1_rad = np.radians(lon1)
    lat2_rad = np.radians(lat2)
    lon2_rad = np.radians(lon2)

    dlon = lon2_rad - lon1_rad
    dlat = lat2_rad - lat1_rad

    a = np.sin(dlat / 2)**2 + np.cos(lat1_rad) * np.cos(lat2_rad) * np.sin(dlon / 2)**2
    c = 2 * np.arctan2(np.sqrt(a), np.sqrt(1 - a))
    distance = R * c

    return distance

# Calculate the distance and create a new column 'Distance_covered'
df_trail['Distance_covered'] = haversine(df_trail['Restaurant_latitude'], df_trail['Restaurant_longitude'],
                                   df_trail['Delivery_location_latitude'], df_trail['Delivery_location_longitude'])
```
